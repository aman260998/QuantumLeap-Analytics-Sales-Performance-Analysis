## **1. Domain Explanation: The World of Sales & Business Analytics**

Before diving into the data, it's crucial to understand the business context. Sales is the engine of any commercial enterprise, and data analytics is the navigation system that ensures this engine is running efficiently and in the right direction.

### **The Business Model: How Sales Operations Work**

Modern sales organizations, especially in the B2B (Business-to-Business) sector, operate on a structured process designed to guide potential clients from initial awareness to a final purchase. This process is often visualized as a **Sales Funnel** or **Sales Pipeline**.

1.  **Lead Generation:** This is the top of the funnel. The company generates leads (potential customers) through various channels like marketing campaigns (`Web`, `Organic`), networking (`Trade Show`), existing relationships (`Referral`), or direct outreach (`Cold Call`).
2.  **Qualification:** The sales team qualifies these leads to determine if they are a good fit for the product—do they have the need, the budget, and the authority to make a purchase?
3.  **Proposal & Negotiation:** For qualified leads, sales representatives (`Sales Reps`) present a tailored proposal. This stage often involves negotiations on price, features, and contract terms. Discounts may be offered to close the deal.
4.  **Closing:** The deal is won, and the prospect becomes a customer. The order is processed, and the logistics of fulfillment begin.
5.  **Retention & Upselling:** The relationship doesn't end at the sale. Successful companies focus on retaining customers and selling them additional products or services over time.

This entire process is managed through different **Sales Channels**. A `Direct` channel involves a sales rep engaging directly with the client. An `Online` channel is self-serve, typically for smaller transactions. `Reseller` or `Partner` channels involve third parties selling the company's products.

### **The Importance of Data in Sales**

In a competitive market, relying on intuition or "gut feelings" is a recipe for failure. Sales data is the source of truth. It provides an objective view of what's working and what isn't. By analyzing transaction data, a company can:
* Identify its most profitable products and customer segments.
* Recognize its top-performing (and underperforming) sales representatives.
* Optimize pricing and discount strategies.
* Forecast future revenue with greater accuracy.
* Understand which sales channels and marketing efforts provide the best return on investment.

Without data analysis, a sales organization cannot effectively scale, adapt to market changes, or hold its teams accountable for their performance.

### **Key Performance Indicators (KPIs) in Sales**

To measure success, sales leaders constantly monitor a set of critical KPIs:

* **Sales Revenue:** The most fundamental metric—the total amount of money generated from sales. This can be broken down by region, product, or sales rep.
* **Quota Attainment:** The percentage of the sales team that is achieving their assigned sales quota (target). This is a key indicator of the team's overall health.
* **Win Rate:** The percentage of qualified leads that are successfully converted into customers. A low win rate might indicate problems with the product, pricing, or sales process.
* **Sales Cycle Length:** The average amount of time it takes to close a deal, from initial contact to the final signature. A long sales cycle can increase costs and risk losing the deal to a competitor.
* **Average Deal Size:** The average revenue generated from each closed deal. A key goal is often to increase this metric through upselling or focusing on larger clients.
* **Lead Conversion Rate:** The percentage of leads from a particular source that become qualified prospects. This helps measure the quality of leads generated by marketing.

### **Common Challenges & The Role of Data Analysis**

Sales organizations face numerous challenges that data analysis can directly address:

* **Challenge: Inaccurate Forecasting.** Predicting future sales is notoriously difficult but essential for financial planning.
    * **Data Analysis Solution:** By analyzing historical sales data, seasonality, and the current sales pipeline, analysts can build statistical models to create more reliable sales forecasts.
* **Challenge: Inconsistent Regional Performance.** Some sales regions may significantly outperform others.
    * **Data Analysis Solution:** A deep-dive analysis can compare regions, identifying differences in product mix, discount strategies, or sales rep performance that explain the disparity.
* **Challenge: Optimizing Discount Strategy.** Offering too many discounts can erode profit margins, while offering too few can lead to lost deals.
    * **Data Analysis Solution:** By analyzing how discounts impact deal size and win rate, a company can identify the "sweet spot" for discounting that maximizes revenue without sacrificing profitability.
* **Challenge: Identifying the Ideal Customer Profile (ICP).** Sales teams can waste significant effort chasing leads that are a poor fit.
    * **Data Analysis Solution:** By analyzing the characteristics of the most profitable and loyal customers (industry, company size, location), analysts can build a data-driven ICP, allowing the sales and marketing teams to focus their efforts on the highest-potential leads.

In short, data analysis empowers a sales organization to make smarter, faster, and more profitable decisions at every stage of the sales cycle.

### **A. Data Cleaning and Preprocessing (Pandas & Core Python)**
1.  Identify all columns with missing values and calculate the percentage of missing data for each. Devise and implement a strategy to handle the NaNs in `SalesRepID`, `CustomerID`, and `LeadSource`.

```python
#Identifying columns with missing values
missing_values=df.isna().sum()
print(f'Number of missing values \n{missing_values[missing_values>0]}')
print('-'*50)
print(f'Percentage of missing values \n{missing_values[missing_values>0]/len(df)*100}')
```

**Reasoning** - Handling missing Customer ID, by matching the Emails of Customers which is also supposed to be unique.

```python
# Handling missing Customer ID
df['CustomerID'] = df['CustomerID'].fillna(df.groupby('CustomerEmail')['CustomerID'].transform('first'))
```

**Reasoning**- Dropping Lead Source Column, because it is completely empty.

```python
# Handling Lead Source column
df.drop(columns='Lead_Source', inplace=True)
```

**Reasoning** -Assigning a Default Id to missing Sales Representative ID, because both the ID and Name of Sales Representative is missing, and there is no column that can uniquely identify sales representative, and the missing values are so large to be dropped.

```python
# Handling missing Sales Representative ID
df['SalesRepID']=df['SalesRepID'].fillna('DefaultID')
```

```python
# After handling missing values
print(f'Missing values after handling:\n{df[['CustomerID','SalesRepID']].isna().sum()}')
```

2.  The `UnitPrice` and `TotalPrice` columns are stored as strings. Write a function to clean these columns by removing '$' and ',', then convert them to a numeric (float) data type.

```python
df['UnitPrice'] = df['UnitPrice'].replace({r'[$,]':''}, regex = True).astype(float)
df['TotalPrice'] = df['TotalPrice'].replace({r'[$,]':''}, regex = True).astype(float)

print('Data types after conversion: ')
print(df[['UnitPrice', 'TotalPrice']].dtypes)
```

3.  The `Quantity` column contains non-numeric text. Clean this column by replacing text with their numeric counterparts (e.g., 'One' -> 1) and handling 'N/A' values (e.g., replacing with the median quantity).

```python
print("Unique values in 'Unique' before replacing and handling missing values:")
print(df['Quantity'].unique())
```

```python
# Replacing text with their numeric counterparts
df['Quantity']=df['Quantity'].replace({
    'One': '1',
    'Two' : '2',
    'Three': '3'
})

# Changing datatype
df['Quantity']=df['Quantity'].astype('float')

# Handling missing values
df['Quantity']=df['Quantity'].fillna(df['Quantity'].median())
```

```python
print("Unique values in 'Quantity' after cleaning:")
print(df['Quantity'].unique())
```

4.  Standardize the `Region` and `ProductCategory` columns to have consistent, clean values.

```python
print("Unique values in Region and Product Category before standardization:\n")
print(df['Region'].unique())
print()
print(df['ProductCategory'].unique())
```

```python
# Standardizing Data
df['Region']=df['Region'].replace({
    'Asia-Pacific': 'APAC',
    'APAC' : 'APAC',
    'N. America': 'North America',
    'EMEA': 'EMEA',
    'Europe':'EMEA',
    'Middle East':'EMEA',
    'Africa':'EMEA',
    'North America':'North America',
    'Europe, Middle East, Africa':'EMEA'
})


df['ProductCategory']=df['ProductCategory'].replace({
    'Furniture': 'Furniture',
    'electronics' : 'Electronics',
    'office_supplies': 'Office Supplies',
    'Software': 'Software',
    'ELEC':'Electronics',
    'Office Supplies':'Office Supplies',
    'Electronics':'Electronics'
})
print("Unique values in Region and Product Category after standardization:\n")
print(df['Region'].unique())
print()
print(df['ProductCategory'].unique())
```

5.  Identify and correct all records where the `ShippingDate` is earlier than the `OrderDate`. What will you do with these dates?

```python
# Identifying incorrect Shipping Dates
inconsistent_rows = df[(df['ShippingDate'] < df['OrderDate'])]
print(f'Inconsistent rows before handling: {len(inconsistent_rows)}')
```

```python
# Handling inconsistency
mask = df['ShippingDate'] < df['OrderDate']

df.loc[mask, 'ShippingDate'] = np.where(
    df.loc[mask, 'OrderStatus'].isin(['Pending', 'Cancelled']),
    np.nan,
    df.loc[mask, 'OrderDate']
)
```

```python
print(f'Inconsistent rows after handling: {(df['ShippingDate'] < df['OrderDate']).sum()}')
```

6.  Create a new column called `Calculated_TotalPrice`. The formula should be `Quantity * UnitPrice * (1 - Discount_Applied)`.

```python
#Handling missing discount values
df['Discount_Applied']=df['Discount_Applied'].fillna(0)

# Creating new column
df['Calculated_TotalPrice']=(df['Quantity']*df['UnitPrice']*(1-df['Discount_Applied'])).round(2)

# Printing first 5 rows of new column
print(f'First few rows of Calculated Total Price column:\n\n{df['Calculated_TotalPrice'].head()}')
```

7.  Create another column `Price_Discrepancy` that shows the difference between the original `TotalPrice` and your new `Calculated_TotalPrice`. How many records have discrepancies?

```python
# Creating price discrepancy column
df['Price_Discrepancy']=df['TotalPrice'] - df['Calculated_TotalPrice']

# Printing first 5 rows of new column
print(f'First few rows of Price Discrepancy column:\n\n{df['Price_Discrepancy'].head()}')
print()
# Number of records having discrepancies
print(f'Number of records having discrepancies: {len(df[df['Price_Discrepancy']!=0])}')
```

8.  Ensure all date columns (`OrderDate`, `ShippingDate`) are converted to the proper datetime format.

```python
print(f'Datatypes before conversion:\n{df[['OrderDate','ShippingDate']].dtypes}')
df['OrderDate']=pd.to_datetime(df['OrderDate'])
print()
df['ShippingDate']=pd.to_datetime(df['ShippingDate'])
print(f'Datatypes before conversion:\n{df[['OrderDate','ShippingDate']].dtypes}')
```

**Recommendation**

Increase marketing investment in Organic and Web lead generation channels.
Expand Direct Sales efforts in high-performing regions such as EMEA.
Adopt a region-specific channel strategy, emphasizing:
Reseller partnerships in EMEA and North America
Online sales channels in APAC

**Expected Impact**

Optimizing channel strategy by region can improve lead quality, conversion rates, and overall revenue growth.

###2. Focus Sales and Marketing Efforts on High-Revenue Products and Categories

Certain products contribute disproportionately to revenue, and categories such as Electronics and Software show the highest average purchase quantities, indicating strong customer demand. In contrast, Furniture has the lowest purchase quantity and the highest return rate (24.6%).

**Recommendation**

Prioritize Electronics and Software products in promotional campaigns and sales initiatives.
Review Furniture product strategy, including pricing, product quality, or product descriptions to reduce return rates.
Introduce bundling or cross-selling strategies combining high-demand products with slower-moving categories.

**Expected Impact**

Focusing on high-demand products while addressing return issues can increase revenue efficiency and improve profit margins.

###3. Use Strategic Discounts and Seasonal Planning to Drive Sales Volume

The analysis shows that orders with discounts have significantly higher average quantities (11.32 vs. 3.06), indicating that discounts strongly influence purchasing behavior. Additionally, revenue peaks during specific months (October, August, January, April, and June).

**Recommendation**

Implement targeted promotional campaigns and discount strategies during historically high-revenue months to maximize sales volume.
Use data-driven discounting, focusing on products where discounts significantly increase order quantities.
Align inventory and marketing planning with seasonal demand patterns.

**Expected Impact**

Strategic discounting and seasonal planning can increase order volume, revenue growth, and inventory turnover.

###Strategic Summary

Overall, revenue growth opportunities lie in three key areas:

Optimizing sales channels and lead acquisition strategies
Prioritizing high-performing products while addressing high-return categories
Leveraging discounts and seasonal demand patterns to maximize order volume

By aligning sales and marketing strategies with these data-driven insights, QuantumLeap Analytics can improve revenue performance, customer acquisition efficiency, and profitability.
